### Difference in Power

```python
from src.spectral import power_change

frequency_of_interest1 = 200
frequency_of_interest2 = 50
time_extent = (-.300, 0.300)
n_trials = 100
n_time_samples = ((time_extent[1] - time_extent[0]) * sampling_frequency) + 1
time = np.linspace(time_extent[0], time_extent[1], num=n_time_samples, endpoint=True)
signal1 = np.sin(2 * np.pi * time * frequency_of_interest1)
signal2 = np.sin(2 * np.pi * time * frequency_of_interest2)
signal2[:int(np.fix(n_time_samples / 2))] = 0 
noise = np.random.normal(0, 2, size=(len(signal1), n_trials))
signal = signal1 + signal2
data = signal1[:,np.newaxis] + signal2[:,np.newaxis] + noise
time_before = time < 0
time_after = time >= 0

fig, axes = plt.subplots(nrows=5, ncols=2, figsize=(15, 20))

axes[0, 0].plot(time, signal1 + signal2)
axes[0, 0].set_ylim((-10, 10))
axes[0, 0].set_xlim(time_extent)
axes[0, 0].set_xlabel('Time')
axes[0, 0].set_ylabel('Amplitude')
axes[0, 0].set_title('Signal', fontweight='bold')
axes[0, 0].axvline(time[int(np.fix(n_time_samples / 2))], color='black')

axes[0, 1].plot(time, data)
axes[0, 1].set_ylim((-10, 10))
axes[0, 1].set_xlim(time_extent)
axes[0, 1].set_xlabel('Time')
axes[0, 1].set_ylabel('Amplitude')
axes[0, 1].set_title('Signal + Noise', fontweight='bold')
axes[0, 1].axvline(time[int(np.fix(n_time_samples / 2))], color='black')

# PSD Before
psd_signal_before = multitaper_power_spectral_density(
    signal[time_before], sampling_frequency, time_halfbandwidth_product=7.5)
psd_signal_before.plot(ax=axes[1, 0])
axes[1, 0].set_ylim((0, 0.012))
axes[1, 0].set_title('PSD -- Time < 0')

psd_data_before = multitaper_power_spectral_density(
    data[time_before, :], sampling_frequency,
    time_halfbandwidth_product=7.5)
psd_data_before.plot(ax=axes[1, 1])
axes[1, 1].set_ylim((0, 0.012))
axes[1, 1].set_title('PSD -- Time < 0')

# PSD After
psd = multitaper_power_spectral_density(
    signal[time_after], sampling_frequency,
    time_halfbandwidth_product=7.5).plot(ax=axes[2, 0])
axes[2, 0].set_ylim((0, 0.012))
axes[2, 0].set_title('PSD -- Time >= 0')

psd = multitaper_power_spectral_density(
    data[time_after, :], sampling_frequency,
    time_halfbandwidth_product=7.5).plot(ax=axes[2, 1])
axes[2, 1].set_ylim((0, 0.012))
axes[2, 1].set_title('PSD -- Time >= 0')

# Spectrogram
spectrogram_signal = multitaper_spectrogram(
    signal, sampling_frequency, time_window_duration=0.080,
    time_window_step=0.040, time_halfbandwidth_product=2, time=time)
plot_spectrogram(spectrogram_signal, axes[3, 0], vmin=0, vmax=0.012)
axes[3, 0].axvline(time[int(np.fix(n_time_samples / 2))], color='black')

spectrogram_data = multitaper_spectrogram(
    data, sampling_frequency, time_window_duration=0.080,
    time_window_step=0.040, time_halfbandwidth_product=2,
    time=time)
plot_spectrogram(spectrogram_data, axes[3, 1], vmin=0, vmax=0.012)
axes[3, 1].axvline(time[int(np.fix(n_time_samples / 2))], color='black')

# Change in power
change_in_power_signal = power_change(psd_signal_before, spectrogram_signal)
plot_spectrogram(change_in_power_signal, axes[4, 0],
                                 vmin=1E-1, vmax=1E1, plot_type='change',
                                 cmap='RdBu_r')
axes[4, 0].axvline(time[int(np.fix(n_time_samples / 2))], color='black')

change_in_power_data = power_change(psd_data_before, spectrogram_data)
mesh = plot_spectrogram(change_in_power_data, axes[4, 1],
                                 vmin=1E-1, vmax=1E1, plot_type='change',
                                 cmap='RdBu_r')
axes[4, 1].axvline(time[int(np.fix(n_time_samples / 2))], color='black')

plt.tight_layout()
cb = fig.colorbar(mesh, ax=axes.ravel().tolist(), orientation='horizontal',
                  shrink=.5, aspect=18, pad=0.1, label='Multiplicative Power Change', format='%.2f')
cb.outline.set_linewidth(0)
cb.set_ticks([1/10, 1/4, 1/3, 1/2, 1, 2, 3, 4, 10])
```
